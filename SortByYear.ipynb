```python
# Import Dependencies
import pandas as pd
import calendar
```

```python
# Import the CSV into a pandas DataFrame
county = "Yuma"
file = county+"_MaxTemps.csv"
try:
    data = pd.read_csv(file, low_memory=False)
    print(f"Read in data ({len(data)})")
except FileNotFoundError:
    print(f"Could not open '{file}'")
```

```python
days = []
for month in range(1,13):
    for day in range(1,calendar.monthrange(2019,month)[1]+1):
        days.append(str(month).zfill(2)+'-'+str(day).zfill(2))
#print(days)
```

```python
df = pd.DataFrame()
df['DAY'] = days
df.head()
```

```python
for row in data.itertuples():
    y = row.Date.split('-',1)[0]
    d = row.Date.split('-',1)[1]
    df.loc[df['DAY']==d, y] = row.Temp
df.head()
```

```python
df.isnull().sum()
```

```python
means = []
for row in range(len(df)):
    means.append(round(df.iloc[row,1:].mean()))
df['AVG'] = means
df.head()
```

```python
df.to_csv(county+"_Yearly.csv", index=False)
```
