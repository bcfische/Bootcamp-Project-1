```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
az_county_suicides = "Resources/az_suicide_rates_by_county.csv"
az_county_deaths = "Resources/az_death_by_year.csv"
az_population = "Resources/az_pop.csv"
az_marriage = "Resources/az_marriage_stats.csv"

suicides_df = pd.read_csv(az_county_suicides, encoding="ISO-8859-1")
death_df = pd.read_csv(az_county_deaths, encoding="ISO-8859-1")
population_df = pd.read_csv(az_population, encoding="ISO-8859-1")
marriage_df = pd.read_csv(az_marriage, encoding="ISO-8859-1")
```

```python
suicides_df.head()
suicides_df.dtypes
```

```python
suicide_new_df = suicides_df.rename(columns={"Unnamed: 0":"County"})
suicide_new_df.head()
```

```python
#suicide_new_df.dtypes
df2 = suicide_new_df.transpose()
df2.columns = df2.iloc[0]
df2.dtypes
```

```python
df2.head(20)
```

```python
df2['ARIZONA'] = pd.to_numeric(df2['ARIZONA'], errors='coerce')
df2['Apache'] = pd.to_numeric(df2['Apache'], errors='coerce')
df2['Cochise'] = pd.to_numeric(df2['Cochise'], errors='coerce')
df2['Coconino'] = pd.to_numeric(df2['Coconino'], errors='coerce')
df2['Gila'] = pd.to_numeric(df2['Gila'], errors='coerce')
df2['Graham'] = pd.to_numeric(df2['Graham'], errors='coerce')
df2['Greenlee'] = pd.to_numeric(df2['Greenlee'], errors='coerce')
df2['La Paz'] = pd.to_numeric(df2['La Paz'], errors='coerce')
df2['Maricopa'] = pd.to_numeric(df2['Maricopa'], errors='coerce')
df2['Mohave'] = pd.to_numeric(df2['Mohave'], errors='coerce')
df2['Navajo'] = pd.to_numeric(df2['Navajo'], errors='coerce')
df2['Pima'] = pd.to_numeric(df2['Pima'], errors='coerce')
df2['Pinal'] = pd.to_numeric(df2['Pinal'], errors='coerce')
df2['Santa Cruz'] = pd.to_numeric(df2['Santa Cruz'], errors='coerce')
df2['Yavapai'] = pd.to_numeric(df2['Yavapai'], errors='coerce')
df2['Yuma'] = pd.to_numeric(df2['Yuma'], errors='coerce')
```

```python
df2.plot(kind="line", figsize=(20,5))

plt.title("Arizona Suicide Rate by County: 2007-2017")
plt.show()
plt.tight_layout()
plt.savefig("Images/counties.png")
```

```python
df2.plot(kind="bar", figsize=(20,5))

plt.title("Arizona Suicide Rate by County: 2007-2017")
plt.show()
plt.tight_layout()
plt.savefig("Images/counties.png")
```

```python
arizona = df2.loc[ : , 'ARIZONA']
arizona.head()

arizona = arizona.rename(columns={"County":"Year"})
```

```python
arizona.plot(kind="line", figsize=(20,5))

plt.title("Arizona Statewide Suicide Rate: 2007-2017")
plt.show()
plt.tight_layout()
plt.savefig("Images/state.png")
```
