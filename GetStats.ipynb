```python
# Import Dependencies
import pandas as pd
```

```python
# Import the CSV into a pandas DataFrame
county = "Apache"
file = county+"_MaxTemps.csv"
try:
    data = pd.read_csv(file, low_memory=False)
    print(f"Read in data ({len(data)})")
except FileNotFoundError:
    print(f"Could not open '{file}'")
```

```python
data['Date'] = data['Date'].str.split('-',1,True)[1]
data.head()
```

```python
grouped_by_day = data.groupby(['Date'])
tcnt = grouped_by_day['Temp'].count()
tavg = grouped_by_day['Temp'].mean()
tmin = grouped_by_day['Temp'].min()
tmax = grouped_by_day['Temp'].max()
daily_stats = pd.DataFrame({
    "Day":[val for val in grouped_by_day.groups.keys()],
    "Count":[val for val in tcnt],
    "Mean":[val for val in tavg],
    "Min":[val for val in tmin],
    "Max":[val for val in tmax]
    })
daily_stats = daily_stats[daily_stats.Day!='02-29']
daily_stats = daily_stats.astype({"Day":str,"Count":int,"Mean":int,"Min":int,"Max":int})
daily_stats.head(70)
```

```python
daily_stats.to_csv(county+"_Stats.csv", index=False)
```
